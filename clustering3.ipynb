```python
# Importing necessary Libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

Scrape data from Wikipedia and turn it into dataframe

```python
# Webpage url                                                                                                               
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

# source of the page's HTML
source = requests.get(url).text
```

```python
soup = BeautifulSoup(source,'xml')

from IPython.display import display_html
tab = str(soup.table)

display_html(tab,raw=True)
print('Page Scrapped.')
```

# convert html to pandas DataFrame


```python

dfs = pd.read_html(tab)
df =dfs[0]
df.head()
```

# Data Wrangling and Cleaning

```python
# dropping the rows where Borough is 'Not assigned'
df = df[df.Borough!= 'Not assigned']

# grouping neighbours with the samr postal code
df = df.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df.reset_index(inplace=True)

df['Neighborhood'] = np.where(df['Neighborhood'] == 'Not assigned',df['Borough'], df['Neighborhood'])

df
```

```python
df.shape
```

# Get coordinates of each postal code

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

```python
df = pd.merge(df,lat_lon,on='Postal Code')
df.head()
```

# Getting all the rows from the data frame which contains Toronto in their Borough

```python
df1 = df[df['Borough'].str.contains('Toronto',regex=False)]
df1
```
